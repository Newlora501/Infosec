1) –í –±–∏–±–ª–∏–æ—Ç–µ–∫–µ Matplotlib –µ—Å—Ç—å —É–Ω–∏–≤–µ—Ä—Å–∞–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è plot, —Å –ø–æ–º–æ—â—å—é –∫–æ—Ç–æ—Ä–æ–π –º–æ–∂–Ω–æ –æ—Ç–æ–±—Ä–∞–∑–∏—Ç—å —Ç–æ—á–∫–∏, –æ—Ç—Ä–µ–∑–∫–∏, –ª—É—á–∏, –ø—Ä—è–º—ã–µ, –ª–æ–º–∞–Ω–Ω—ã–µ, –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∏, –∫—Ä–∏–≤—ã–µ –∏ —Ç.–¥. –û–î–Ω–∞–∫–æ, –∏–∑-–∑–∞ –µ–µ —É–Ω–∏–≤–µ—Ä—Å–∞–ª—å–Ω–æ—Å—Ç–∏ –¥–µ–ª–∞—Ç—å —ç—Ç–æ –∑–∞—á–∞—Å—Ç—É—é –Ω–µ —É–¥–æ–±–Ω–æ, –ø–æ—ç—Ç–æ–º—É –ª—É—á—à–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å–ø–µ—Ü–∏–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏.
‚Ä¢ –ó–∞–¥–∞–π—Ç–µ –¥–∏–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ –æ—Å—è–º $ ùëÇùë•ùë¶ $ –∫–∞–∫ $[0, ùëõ] √ó [0, ùëõ]$, –≥–¥–µ $ùëõ$ ‚Äî –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ.

‚Ä¢ –ù–∞—Å—Ç—Ä–æ–π—Ç–µ –≤–∏–¥ –ª–∏–Ω–∏–π, –∏–∑–º–µ–Ω–∏–≤ —Ü–≤–µ—Ç, —Å—Ç–∏–ª—å (–ø—É–Ω–∫—Ç–∏—Ä, —à—Ç—Ä–∏—Ö–ø—É–Ω–∫—Ç–∏—Ä, —à—Ç—Ä–∏—Ö –∏ —Ç.–¥.), —Ç–æ–ª—â–∏–Ω—É –∏ —Ç.–¥.
‚Ä¢ –ò–∑—É—á–∏—Ç–µ —Å–ø—Ä–∞–≤–∫—É —Ñ—É–Ω–∫—Ü–∏–π axvline –∏ axhline. –ß–µ–º –æ–Ω–∏ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è –æ—Ç –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö –∏ –º–æ–∂–Ω–æ –ª–∏

‚Ä¢ –ò–∑—É—á–∏—Ç–µ —Å–ø—Ä–∞–≤–∫—É —Ñ—É–Ω–∫—Ü–∏–∏ axline. –ó–∞ —á—Ç–æ –æ—Ç–≤–µ—á–∞–µ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä slope? –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —ç—Ç—É —Ñ—É–Ω–∫—Ü–∏—é,
—á—Ç–æ–±—ã –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å –æ–±–µ –¥–∏–∞–≥–æ–Ω–∞–ª–∏ –≤—Å–µ—Ö –ø–æ–ª—É—á–∏–≤—à–∏—Ö—Å—è –∫–≤–∞–¥—Ä–∞—Ç–æ–≤.
‚Ä¢ –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ —Ä–∞–±–æ—Ç–∞—Ç—å —Ç–∞–∫, —á—Ç–æ–±—ã –ø—Ä–∞–≤–∏–ª—å–Ω–æ —Ä–∏—Å–æ–≤–∞—Ç—å —Å–µ—Ç–∫—É –ø—Ä–∏ –ª—é–±–æ–º –∑–∞–¥–∞–Ω–Ω–æ–º —Ü–µ–ª–æ–º ùëõ.

–§—É–Ω–∫—Ü–∏–∏ axvline –∏ axhline —è–≤–ª—è—é—Ç—Å—è —Å–ø–µ—Ü–∏—Ñ–∏—á–Ω—ã–º–∏ –¥–ª—è –æ–±—ä–µ–∫—Ç–æ–≤ Axes —Ñ—É–Ω–∫—Ü–∏—è–º–∏ Matplotlib –∏ –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è –¥–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã—Ö (vlines) –∏ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—ã—Ö (hlines) –ª–∏–Ω–∏–π –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ. –ò—Ö –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤—ã —Ñ—É–Ω–∫—Ü–∏—è–º vlines –∏ hlines –ø–æ–¥–º–æ–¥—É–ª—è pyplot.

–ö–ª—é—á–µ–≤–æ–µ —Ä–∞–∑–ª–∏—á–∏–µ –º–µ–∂–¥—É —ç—Ç–∏–º–∏ –¥–≤—É–º—è –ø–æ–¥—Ö–æ–¥–∞–º–∏ –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤ —Ç–æ–º, —á—Ç–æ —Ñ—É–Ω–∫—Ü–∏–∏ axvline –∏ axhline —Ä–∞–±–æ—Ç–∞—é—Ç –Ω–µ–ø–æ—Å—Ä–µ–¥—Å—Ç–≤–µ–Ω–Ω–æ —Å –æ–±—ä–µ–∫—Ç–æ–º Axes, —Ç–æ–≥–¥–∞ –∫–∞–∫ —Ñ—É–Ω–∫—Ü–∏–∏ vlines –∏ hlines —è–≤–ª—è—é—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏—è–º–∏ —É—Ä–æ–≤–Ω—è –º–æ–¥—É–ª—è, –∫–æ—Ç–æ—Ä—ã–µ —Ä–∞–±–æ—Ç–∞—é—Ç —Å–æ –≤—Å–µ–π —Ñ–∏–≥—É—Ä–æ–π. –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ axvline –∏ axhline –¥–∞–µ—Ç –≤–∞–º –±–æ–ª—å—à–µ –∫–æ–Ω—Ç—Ä–æ–ª—è –Ω–∞–¥ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏–µ–º –ª–∏–Ω–∏–π –∏ –æ—Å–æ–±–µ–Ω–Ω–æ –ø–æ–ª–µ–∑–Ω–æ, –∫–æ–≥–¥–∞ –≤—ã —Ä–∞–±–æ—Ç–∞–µ—Ç–µ —Å –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ –ø–æ–¥–≥—Ä–∞—Ñ–∞–º–∏ (–ø–æ–¥–∑–∞–≥–æ–ª–æ–≤–∫–∞–º–∏) –Ω–∞ –æ–¥–Ω–æ–π —Ñ–∏–≥—É—Ä–µ.

```python
n = 15 #–≤—ã–±–µ—Ä–∏—Ç–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ n
```

```python
def fig():
  fig01= plt.figure(num=1, figsize=(5,5), dpi=200)
  return
fig()
plt.xlim(0, n)
plt.ylim(0, n)

# –û—Ç–∫–ª—é—á–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω—É—é —Å–µ—Ç–∫—É
plt.grid(False)

# –†–∞–∑–º–µ—Ä –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã—Ö –∫–ª–µ—Ç–æ–∫
square_size = 1

# –ù–∞—Ä–∏—Å—É–π—Ç–µ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–µ –∫–ª–µ—Ç–∫–∏ —Å–æ —Å—Ç–æ—Ä–æ–Ω–æ–π 1, –∑–∞–º–æ—Å—Ç–∏–≤ –∏–º–∏ –≤—Å—é –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω—É—é –ø–ª–æ—Å–∫–æ—Å—Ç—å.

def grid() :
  for x in range(n+1):
    plt.axvline(x, color='gray', linestyle='-', lw=1)  # –í–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏
    plt.axhline(x, color='gray', linestyle='-', lw=1)  # –ì–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏
  return


# –ò–∑—É—á–∏—Ç–µ —Å–ø—Ä–∞–≤–∫—É —Ñ—É–Ω–∫—Ü–∏–π axvline –∏ axhline. –≠—Ç–∏ —Ñ—É–Ω–∫—Ü–∏–∏ –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã—Ö
# –∏ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—ã—Ö –ª–∏–Ω–∏–π —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ. –û–Ω–∏ –º–æ–≥—É—Ç –±—ã—Ç—å –ø–æ–ª–µ–∑–Ω—ã –¥–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è —Å–µ—Ç–∫–∏, –Ω–æ
# –¥–ª—è –¥–∞–Ω–Ω–æ–π –∑–∞–¥–∞—á–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ axline –º–æ–∂–µ—Ç –±—ã—Ç—å –±–æ–ª–µ–µ —É–¥–æ–±–Ω—ã–º.

# –ò–∑—É—á–∏—Ç–µ —Å–ø—Ä–∞–≤–∫—É —Ñ—É–Ω–∫—Ü–∏–∏ axline. –ü–∞—Ä–∞–º–µ—Ç—Ä slope –æ—Ç–≤–µ—á–∞–µ—Ç –∑–∞ –Ω–∞–∫–ª–æ–Ω –ª–∏–Ω–∏–∏.
# –ú—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —ç—Ç—É —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è –¥–∏–∞–≥–æ–Ω–∞–ª–µ–π –∫–≤–∞–¥—Ä–∞—Ç–æ–≤.

# –ù–∞—Ä–∏—Å—É–π—Ç–µ –¥–∏–∞–≥–æ–Ω–∞–ª–∏ –≤—Å–µ—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤.
for i in range(n):
    slope = 1  # –ù–∞–∫–ª–æ–Ω –¥–∏–∞–≥–æ–Ω–∞–ª–∏
    plt.axline((i, i), slope=slope, color='blue', linestyle='--', lw=1)  # –ì–ª–∞–≤–Ω–∞—è –¥–∏–∞–≥–æ–Ω–∞–ª—å
    plt.axline((i, n - i), slope=-slope, color='red', linestyle='--', lw=1)  # –ü–æ–±–æ—á–Ω–∞—è –¥–∏–∞–≥–æ–Ω–∞–ª—å

# –ü–æ–∫–∞–∂–∏—Ç–µ –≥—Ä–∞—Ñ–∏–∫
grid()
plt.show()
```

2) –î–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–æ–≤ –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ä–∞–∑–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π,–Ω–∞–ø—Ä–∏–º–µ—Ä —Ñ—É–Ω–∫—Ü–∏—é arrow –∏–ª–∏ annotate.
‚Ä¢ –ù–∞—Ä–∏—Å—É–π—Ç–µ –¥–≤–∞ –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–∞. –ù–∞–π–¥–∏—Ç–µ –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —É–≥–æ–ª –º–µ–∂–¥—É –Ω–∏–º–∏.–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é $\arctan2$ –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ NumPy –∏–ª–∏ atan2 –∏–∑ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –º–æ–¥—É–ª—è math. –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç—å –≤—ã—á–∏—Å–ª–µ–Ω–∏–π –Ω–∞ –ø—Ä–∏–º–µ—Ä–∞—Ö –æ—Å—Ç—Ä–æ–≥–æ, —Ç—É–ø–æ–≥–æ –∏ –ø—Ä—è–º–æ–≥–æ —É–≥–ª–æ–≤.
‚Ä¢ –ü—Ä–∏–º–µ–Ω–∏—Ç–µ –∫ –≤–µ–∫—Ç–æ—Ä–∞–º —Ç–∞–∫–æ–µ –ª–∏–Ω–µ–π–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ, —á—Ç–æ–±—ã –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —É–≥–æ–ª –ø–æ–º–µ–Ω—è–ª —Å–≤–æ–π –∑–Ω–∞–∫.
‚Ä¢ –ù–∞—Ä–∏—Å—É–π—Ç–µ —Ç–æ –∂–µ —Å–∞–º–æ–µ –ø–µ—Ä–µ–∫–ª—é—á–∏–≤—à–∏—Å—å –≤ –ø–æ–ª—è—Ä–Ω—É—é —Å–∏—Å—Ç–µ–º—É –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç.

```python

vector1 = np.array([2, 3])
vector2 = np.array([1, 4])

# Calcule o √¢ngulo entre os dois vetores usando a fun√ß√£o arctan2 do NumPy
angle_radians = np.arctan2(np.linalg.det([vector1, vector2]), np.dot(vector1, vector2))

# Converta o √¢ngulo de radianos para graus
angle_degrees = np.degrees(angle_radians)

# Crie uma figura e um eixo para o gr√°fico
fig, ax = plt.subplots(num=1, figsize=(5,5), dpi=200)

# Desenhe os vetores no gr√°fico
ax.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector 1')
ax.quiver(0, 0, vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector 2')

# Configure os limites do gr√°fico
ax.set_xlim(0, n)
ax.set_ylim(0, n)

# Adicione r√≥tulos e uma legenda
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.legend()

# Adicione o √¢ngulo entre os vetores no gr√°fico
ax.annotate(f'√Çngulo = {angle_degrees:.2f} graus', xy=(0.5, 0.1), xycoords='axes fraction')

# Exiba o gr√°fico
plt.grid(False)
grid()

plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
vector2_inverted = -vector2

# Calcule o √¢ngulo entre os vetores invertidos
angle_radians_inverted = np.arctan2(np.linalg.det([vector1, vector2_inverted]), np.dot(vector1, vector2_inverted))
angle_degrees_inverted = np.degrees(angle_radians_inverted)

# Imprima o √¢ngulo invertido
print(f'√Çngulo Invertido = {angle_degrees_inverted:.2f} graus')
```

```python
# Crie uma figura polar
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'},num=1, figsize=(5,5), dpi=200)

# Calcule o m√≥dulo e o √¢ngulo dos vetores
r1 = np.linalg.norm(vector1)
theta1 = np.arctan2(vector1[1], vector1[0])
r2 = np.linalg.norm(vector2)
theta2 = np.arctan2(vector2[1], vector2[0])

# Desenhe os vetores no gr√°fico polar
ax.quiver(theta1, r1, 0.1, 0, angles='xy', scale_units='xy', scale=1, color='blue', label='Vector 1')
ax.quiver(theta2, r2, 0.1, 0, angles='xy', scale_units='xy', scale=1, color='red', label='Vector 2')

# Adicione uma legenda
ax.legend()

# Exiba o gr√°fico polar
plt.show()
```

‚Ä¢ –ö–∞–∫ –∏–∑ –ø–æ–ª—É—á–∏–≤—à–µ–≥–æ—Å—è –º–∞—Å—Å–∏–≤–∞ –≤—ã–±—Ä–∞—Ç—å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã $x$ –≤—Å–µ—Ö –≤–µ—Ä—à–∏–Ω?–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã $y$? –ö–∞–∫ –ø–æ–ª—É—á–∏—Ç—å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã $(x, y)$ –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–π $i$-–æ–π –≤–µ—Ä—à–∏–Ω—ã?‚Ä¢ –î–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é Polygon –∏–∑ –ø–æ–¥–º–æ–¥—É–ª—è Matplotlib.patches,
–∞ –¥–ª—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ ‚Äî —Ñ—É–Ω–∫—Ü–∏—é RegularPolygon.
‚Ä¢ –í—ã–¥–µ–ª–∏—Ç–µ –≤–µ—Ä—à–∏–Ω—ã –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ –º–∞—Ä–∫–µ—Ä–Ω—ã–º–∏ —Ç–æ—á–∫–∞–º–∏.–ö–∞–∫ –Ω–∞—Å—Ç—Ä–æ–∏—Ç—å —Ä–∞–∑–º–µ—Ä –º–∞—Ä–∫–µ—Ä–æ–≤? –¶–≤–µ—Ç?

```python
from matplotlib.patches import Polygon
```

```python

# Defina as coordenadas dos v√©rtices do pol√≠gono como uma matriz de pontos (x, y)
vertices = np.array([[4, 1], [10, 1], [10, 8], [4, 8]])

# Crie uma figura e um eixo para o gr√°fico
fig, ax = plt.subplots(num=1, figsize=(5,5), dpi=200)

# Desenhe o pol√≠gono com as coordenadas especificadas
polygon = Polygon(vertices, closed=True, fill=False, edgecolor='b', lw=2)
ax.add_patch(polygon)

# Defina o tamanho e a cor dos marcadores dos v√©rtices
marker_size = 8
marker_color = 'r'

# Desenhe os v√©rtices com marcadores
ax.scatter(vertices[:, 0], vertices[:, 1], s=marker_size, c=marker_color, marker='o', label='V√©rtices')

# Calcule a √°rea orientada do pol√≠gono usando a f√≥rmula de Shoelace
def oriented_area(vertices):
    n = len(vertices)
    area = 0
    for i in range(n):
        x1, y1 = vertices[i]
        x2, y2 = vertices[(i + 1) % n]
        area += (x1 * y2 - x2 * y1)
    return 0.5 * abs(area)

area = oriented_area(vertices)

# Adicione r√≥tulos e uma legenda
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title(f'Pol√≠gono com √Årea Orientada = {area:.2f}')
ax.legend()

# Configure os limites do gr√°fico
ax.set_xlim(0, n)
ax.set_ylim(0, n)

# Exiba o gr√°fico
plt.grid(False)
grid()
plt.show()
```

1) –ù–∞—Ä–∏—Å—É–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ —Ü–∏–∫–ª–æ–∏–¥—ã –¥–ª—è –¥–≤—É—Ö –ø–æ–ª–Ω—ã—Ö –æ–±–æ—Ä–æ—Ç–æ–≤ –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏. –û—Ç–æ–±—Ä–∞–∑–∏—Ç–µ –ø–æ–ª–µ –∫–∞—Å–∞—Ç–µ–ª—å–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ –∫ –¥–∞–Ω–Ω–æ–º—É –≥—Ä–∞—Ñ–∏–∫—É, —Å–º —Ä–∏—Å. 2. –£—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ü–∏–∫–ª–æ–∏–¥—ã:
$$
r(t)=
\left(\begin{array}{cc}
x(t) \\
y(t)
\end{array}\right)
=\begin{cases}
r(t-\sin(t))\\
r(1-\cos(t))
\end{cases}
$$

```python

# Entering the initial data:
r=3
t = np.linspace(-2*np.pi, 2*np.pi, 100)
# initialize the function:
x = r*(t-np.sin(t))
y = r*(1-np.cos(t))
# take the derevative:
dx = r-r*np.cos(t)
dy = r*np.sin(t)

fig = plt.figure(num=3, dpi=200)
ax = fig.add_subplot(1, 1, 1)

# Rendering the circle:
l = np.linspace(0, 2 * np.pi, 100)
X_circle = [(r * np.cos(i) + x[25]) for i in l]
Y_circle = [(r * np.sin(i) + r) for i in l]
ax.plot(X_circle, Y_circle, marker="", color="red", linewidth=0.6)
ax.plot(x[25], r, marker=".", ms=3, color="blue")\

# Rendering the tangent field and the cycloid: (with quiver!!!)
ax.quiver(x[::4], y[::4], dx[::4], dy[::4], units='xy', angles='xy', width=0.07, scale=2.0, label='Tangent vectors')
ax.plot(x, y, linewidth=1, color='black', label=r'cycloid')

ax.set_xlabel(r'x $Ox$')
ax.set_ylabel(r'y $Oy$')
ax.legend(loc=1)
ax.set_aspect('equal')
```

```python

t = np.linspace(0, 2*np.pi, 1000)

# Calcule as coordenadas x e y da astroide para um valor espec√≠fico de a
a = 1
x = a * np.cos(t)**3
y = a * np.sin(t)**3

# Calcule as derivadas de primeira e segunda ordem
x_prime = -3 * a * np.cos(t)**2 * np.sin(t)
y_prime = 3 * a * np.cos(t) * np.sin(t)**2
x_double_prime = -3 * a * (np.cos(t)**3 - 2 * np.sin(t)**2 * np.cos(t))
y_double_prime = 3 * a * (2 * np.cos(t)**2 * np.sin(t) - np.sin(t)**3)

# Calcule o raio de curvatura em um ponto arbitr√°rio (por exemplo, t = np.pi/4)
t_point = np.pi/4
R = np.abs((x_prime*y_double_prime - y_prime*x_double_prime)**(3/2)) / (x_prime**2 + y_prime**2)**(3/2)

# Crie uma figura e um eixo para o gr√°fico
fig, ax = plt.subplots(num=1, figsize=(5,5), dpi=200)

# Plote a astroide
ax.plot(x, y, label='Astroide')

# Plote um ponto na astroide com o raio de curvatura
x_point = a * np.cos(t_point)**3
y_point = a * np.sin(t_point)**3
ax.scatter([x_point], [y_point], color='red', label='Ponto na Astroide')


# Adicione r√≥tulos e uma legenda
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.legend()

# Exiba o gr√°fico

plt.show()
```
